```python
from google.colab import files\nfrom google.colab.patches import cv2_imshow\nimport cv2\nimport numpy as np\n\nuploaded = files.upload()\nimage_path = next(iter(uploaded))\n\nimage = cv2.imread(image_path)\nif image is None:\n    raise ValueError('Could not read the uploaded image.')\n\ngray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)\ngray = cv2.equalizeHist(gray)\n\ncascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'\nface_cascade = cv2.CascadeClassifier(cascade_path)\n\nfaces = face_cascade.detectMultiScale(\n    gray,\n    scaleFactor=1.08,\n    minNeighbors=4,\n    minSize=(35, 35),\n)\n\nresult = image.copy()\nfor x, y, w, h in faces:\n    cv2.rectangle(result, (x, y), (x + w, y + h), (0, 200, 0), 3)\n\nprint(f'Detected {len(faces)} face(s).')\ncv2_imshow(result)
```
